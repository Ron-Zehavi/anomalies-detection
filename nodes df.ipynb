```python
def check_year(data, year=2020):
    """
    :param data: dictionary of the data set
    :return: DataFrame of the data set
    """
    
    df = pd.DataFrame.from_dict(data['samplings'], orient='index')
    df['DateTime'] = pd.to_datetime(df['DateTime'], unit='s')
    df = df.set_index('DateTime')
    df.drop(df.index[0], inplace=True)
    df = df[df.index.year >= year]
    
    return df.shape[0] > 0
```

```python
for node in tqdm_notebook(range(6000)):
    try:
        point = client.get_all(point_id=node, start=2020)
        cur_sector = point['point']['group']['sector']['id']
        sector_name = point['point']['group']['sector']['name']
        cur_authority = point['point']['group']['water_authority']['id']
        authority_name = point['point']['group']['water_authority']['name']
        
        if point['point']['group']['water_authority']['id'] in water_authority:
            if check_year(point):
                if (cur_sector, sector_name) not in sector_dict:
                    sector_dict[(cur_sector, sector_name)] = [node]   
                else:
                    temp_list = sector_dict[(cur_sector, sector_name)]
                    temp_list.append(node)
                    sector_dict[(cur_sector, sector_name)] = temp_list
            
                if (cur_authority, authority_name) not in water_authority_dict:
                    water_authority_dict[(cur_authority, authority_name)] = [node]   
                else:
                    temp_list = sector_dict[(cur_authority, authority_name)]
                    temp_list.append(node)
                    water_authority_dict[(cur_authority, authority_name)] = temp_list
                    
        else:
            print('foriegn water authority')
            
    except:
        print(f'no node in {node}')
        pass
```
